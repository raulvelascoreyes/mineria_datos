### 3.2 Normalizaci칩n de Datos

**Instrucciones:** Aunque la Regresi칩n Log칤stica no es tan sensible a la escala como K-Means, normalizar los datos puede mejorar la convergencia del algoritmo de optimizaci칩n.

````python
# TODO: Verificar el rango de valores de los p칤xeles
print(f"Valor m칤nimo en X: {X.min()}")
print(f"Valor m치ximo en X: {X.max()}")
print(f"\nLos valores ya est치n en el rango [0, 16]")

# Para Regresi칩n Log칤stica, es com칰n normalizar a [0, 1]
# TODO: Normalizar los datos dividiendo entre el valor m치ximo
X_train_norm = X_train / 16.0
X_test_norm = X_test / 16.0

print(f"\nDespu칠s de normalizaci칩n:")
print(f"Valor m칤nimo: {X_train_norm.min()}")
print(f"Valor m치ximo: {X_train_norm.max()}")
````

## Parte 4: Entrenamiento del Modelo

### 4.1 Modelo Baseline: Regresi칩n Log칤stica Simple

**Instrucciones:** Entrena una Regresi칩n Log칤stica con par치metros por defecto.

````python
# TODO: Crear y entrenar el modelo de Regresi칩n Log칤stica
# Usa max_iter=10000 para asegurar convergencia
modelo_baseline = LogisticRegression(max_iter=10000, random_state=42)

# Entrenar el modelo
print("Entrenando modelo baseline...")
modelo_baseline.fit(X_train_norm, y_train)
print("九 Modelo entrenado exitosamente")
````

### 4.2 Realizar Predicciones

**Instrucciones:** Usa el modelo entrenado para hacer predicciones en el conjunto de prueba.

````python
# TODO: Hacer predicciones en el conjunto de prueba
y_pred_baseline = modelo_baseline.predict(X_test_norm)

# Mostrar algunas predicciones
print("Primeras 10 predicciones:")
print(f"Predicciones: {y_pred_baseline[:10]}")
print(f"Valores reales: {y_test[:10]}")
````

### 8.4 Evaluaci칩n en Datos de Entrenamiento vs. Prueba

**Instrucciones:** Compara el desempe침o de cada modelo en train vs. test para detectar sobreajuste.

````python
# TODO: Calcular accuracy en train y test para cada modelo
modelos = {
    'Baseline (C=1)': modelo_baseline,
    'Ridge (C=0.01)': modelo_ridge,
    'Flexible (C=100)': modelo_flexible
}

resultados_train_test = []

for nombre, modelo in modelos.items():
    # Accuracy en train
    y_train_pred = modelo.predict(X_train_norm)
    acc_train = accuracy_score(y_train, y_train_pred)
    
    # Accuracy en test
    y_test_pred = modelo.predict(X_test_norm)
    acc_test = accuracy_score(y_test, y_test_pred)
    
    # Diferencia (gap)
    gap = acc_train - acc_test
    
    resultados_train_test.append({
        'Modelo': nombre,
        'Accuracy Train': acc_train,
        'Accuracy Test': acc_test,
        'Gap': gap
    })

df_train_test = pd.DataFrame(resultados_train_test)

print("\n" + "="*90)
print("游늵 AN츼LISIS DE SOBREAJUSTE (OVERFITTING)")
print("="*90)
print(df_train_test.to_string(index=False))
print("="*90)
print("\n游눠 Nota: Un gap grande indica posible sobreajuste")
````
